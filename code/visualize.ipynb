```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded file
file_path = 'node_metrics.xlsx'
node_metrics_df = pd.read_excel(file_path)

# Generate histograms for each numeric column
numeric_columns = node_metrics_df.select_dtypes(include='number').columns

for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(node_metrics_df[column].dropna(), bins=20, alpha=0.75, edgecolor='black')
    plt.title(f"Histogram for {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
# Display the first few rows to verify the content
node_metrics_df.head()

# Generate histograms for each column and organize them in a matrix plot
num_columns = len(node_metrics_df.columns)
fig, axes = plt.subplots(nrows=(num_columns + 1) // 2, ncols=2, figsize=(12, num_columns * 3))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Generate histograms for each column
for i, column in enumerate(node_metrics_df.columns):
    axes[i].hist(node_metrics_df[column], bins=20, alpha=0.7, color='blue', edgecolor='black')
    axes[i].set_title(f"Histogram of {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Frequency")

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Skip the first column when generating histograms
columns_to_plot = node_metrics_df.columns[1:]  # Exclude the first column

# Adjust the number of rows and columns for the plot
num_columns = len(columns_to_plot)
fig, axes = plt.subplots(nrows=(num_columns + 1) // 2, ncols=2, figsize=(12, num_columns * 3))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Generate histograms for each column (excluding the first column)
for i, column in enumerate(columns_to_plot):
    axes[i].hist(node_metrics_df[column], bins=20, alpha=0.7, color='blue', edgecolor='black')
    axes[i].set_title(f"Histogram of {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Frequency")

# Remove unused subplots
for j in range(i, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
